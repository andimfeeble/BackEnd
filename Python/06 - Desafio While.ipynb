```python
#Exercicio59
from time import sleep
num1 = int(input('Digite o primeiro numero da operação: '))
num2 = int(input('Digite o segundo numero da operação: '))
opcao = 0
while opcao != 5:
    print('''
    [ 1 ] Somar
    [ 2 ] Multiplicar
    [ 3 ] Maior
    [ 4 ] Novos números
    [ 5 ] Sair do programa
    ''')
    opcao = int(input('Digite a opção desejada:'))
    if opcao == 1:
        soma = num1 + num2
        print(f'A soma entre {num1} + {num2} = {soma}')
    elif opcao == 2:
        produto = num1 * num2
        print(f'A multiplicação entre {num1} x {num2} = {produto}')
    elif opcao == 3:
        if num1 > num2:
            print(f'O numero {num1} é maior que o numero {num2}')
        if num2 > num1:
            print(f'O numero {num2} é maior que o numero {num1}')
        else:
            print('Os dois numeros digitados são iguais!')
    elif opcao == 4:
        num1 = int(input('Digite o primeiro numero da operação: '))
        num2 = int(input('Digite o segundo numero da operação: '))
    elif opcao == 5:
        print('Finalizando...')
        sleep(1)
    else:
        print('Opção invalida!')
print('Fim do Programa!')
```

```python
#Exercicio68
from random import randint
soma = cont = 0
while True:
    comp = randint(1, 10)
    opcao = str(input('PAR ou IMPAR: ')).upper().strip()[0]
    if opcao in 'PI':
        user = int(input('Agora, escolha um numero de 1 a 10: '))
        soma = comp + user
        if soma % 2 == 0:
            resultado = 'P'
            if opcao == resultado:
                cont = cont + 1
                print(f'Você Ganhou!!!')
            else:
                print(f'Você Perdeu!!!')
                break
        else:
            resultado = 'I'
            if opcao == resultado:
                cont = cont + 1
                print(f'Você Ganhou!!!')
            else:
                print(f'Você Perdeu!!!')
                break
        print(f'Computador: {comp} -> Usuário: {user} -> Soma: {soma}!')
    else:
        print('Opção Inválida!')
print(f'Computador: {comp} -> Usuário: {user} -> Soma: {soma}!')
print(f'Fim do Jogo, numero de acertos: {cont}')
```

```python
#Exercicio69
mulher = homem = maior = menor = mulher_menor = 0
pessoa = 1
while True:
#   _____________________Cadastro da idade da pessoa_______________________
    idade = int(input(f'Digite a idade da {pessoa}ª pessoa: '))
    if idade < 18:
        menor += 1
    else:
        maior += 1
#   _____________________Cadastro do sexo da pessoa_______________________
    sexo = ' '
    while sexo not in 'FM':
        sexo = str(input(f'Digite [M/F] para o sexo da {pessoa}ª pessoa: ')).upper().strip()[0]
    if sexo == 'F':
        mulher += 1
    if sexo == 'M':
        homem += 1
    if sexo == 'F' and idade < 20:
        mulher_menor +=1
    print(f'{pessoa}º Cadastro -> Idade: {idade} anos -> Sexo: {sexo}.')
    print('-' * 30)
    pessoa += 1
#   _____________________Condição para continuar_______________________
    opcao = ' '
    while opcao not in 'SN':
        opcao = str(input(f'Digite [S/N] se deseja cadastrar outra pessoa: ')).upper().strip()[0]
    if opcao == 'N':
        break
print('Cadastros finalizados com sucesso!')
print('-' * 30)
print(f'Homens cadastrados: {homem} -> Mulheres cadastradas: {mulher}.')
print(f'Maiores de idade: {maior} -> Menores de idade: {menor}.')
print(f'Mulheres com menos de 20 anos cadastradas: {mulher_menor}.')
```
