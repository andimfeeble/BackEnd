```python
#Exercicio78
valores = []
maior = menor = 0
for indice in range(0, 5):
    valores.append(int(input(f'Digite um valor para o indice {indice}: ')))
    if indice == 0:
        maior = menor = valores[indice]
    else:
        if valores[indice] > maior:
            maior = valores[indice]
        if valores[indice] < menor:
            menor = valores[indice]
print(f'Os valores digitados foram: {valores}')
print(f'O maior valor digitado foi: {maior}, nos indices: ', end = '')
for indice, valor in enumerate(valores):
    if valor == maior:
        print(f'{indice}', end = ' ')
print(f'\nO menor valor digitado foi: {menor}, nos indices: ', end = '')
for indice, valor in enumerate(valores):
    if valor == menor:
        print(f'{indice}', end = ' ')
```

```python
#Exercicio80
valores = []
for indice in range(0, 5):
    novo = int(input('Digite um número: '))
    if indice == 0 or novo > valores[-1]:
        valores.append(novo)
    else:
        verifica = 0
        while verifica < len(valores):
            if novo <= valores[verifica]:
                valores.insert(verifica, novo)
                break
            verifica += 1
print(f'Os valores digitados foram: {valores}')
```

```python
#Exercicio83
expressao = str(input('Digite a expressão que deseja analisar: '))
abre_chave = fecha_chave = 0
for caracter in expressao:
    if caracter == '(':
        abre_chave += 1
    if caracter == ')':
        fecha_chave += 1
if abre_chave == fecha_chave:
    print(f'A expressão {expressao} está correta!')
else:
    print(f'a expressão {expressao} está incorreta!')
```

```python
#Exercicio84
listagem = []
maior = []
menor = []
opcao = ''
while opcao != 'N':
    nome = str(input('Digite o nome da pessoa: ')).strip().capitalize()
    peso = int(input('Digite o peso da pessoa: '))
    listagem.append([nome, peso])
    opcao = str(input('Deseja continuar? [S/N] ')).strip().upper()
for indice in range(0, len(listagem)):
    if indice == 0:
        maior.append(listagem[indice])
        menor.append(listagem[indice])
    else:
        for pessoa in maior:
            if pessoa[1] < listagem[indice][1]:
                maior.clear()
                maior.append(listagem[indice])
        for pessoa in menor:
            if pessoa[1] > listagem[indice][1]:
                menor.clear()
                menor.append(listagem[indice])
print(f'Foram cadastradas {len(listagem)} pessoas.')
print(f'A pessoa mais pesada é: {maior}')
print(f'A pessoa mais leve é: {menor}')
```

```python
#Exercicio88
from random import randint
from time import sleep
quantidade = int(input('Quantos jogos deseja gerar? '))
jogos = []
jogo = []
print('-----------MEGA-SENA------------')
print('=' * 32)
for linha in range(0, quantidade):
    for coluna in range(0, 6):
        num_sort = randint(1, 60)
        for num_jogo in jogo:
            if num_sort == num_jogo:
                num_sort = randint(1, 60)
            else:
                pass
        jogo.append(num_sort)
    jogo.sort()
    jogos.append(jogo[:])
    jogo.clear()
    print(f'{linha + 1}º JOGO: {jogos[linha]}')
    sleep(1)
print('=' * 32)
```

```python
#Exercicio89
opcao = ''
lista = []
print('-------CADASTRO DE ALUNO--------')
while opcao != 'N':
    nome = str(input('Digite o nome do aluno: ')).strip().capitalize()
    nota1 = float(input('Digite a primeira nota: '))
    nota2 = float(input('Digite a segunda nota: '))
    media = (nota1 + nota2) / 2
    lista.append([nome, nota1, nota2, media])
    print('-------Cadastro Realizado-------')
    print()
    opcao = str(input('Deseja cadastrar outro aluno? [S/N] ')).strip().upper()
print()
print('--------BOLETIM DA TURMA---------')
print(f'{"Cod.":<5} {"NOME":<10} {"MEDIA":>10}')
print('-' * 32)
for indice, aluno in enumerate(lista):
    print(f'{indice + 1:<5} {aluno[0]:<10} {aluno[3]:>10.1f}')
opcao = ''
print('-' * 32)
print()
while opcao != 'N':
    codigo = int(input('Digite o cod. do Aluno para consulta: '))
    print(f'{"Cod.":<5} {"NOME":<10} {"1ª NOTA":>10} {"2ª NOTA":>10}')
    if codigo >= 1 and codigo <= len(lista):
        for indice, aluno in enumerate(lista):
            if (codigo - 1) == indice:
                print('-' *38)
                print(f'{indice + 1:<5} {aluno[0]:<10} {aluno[1]:>10.1f} {aluno[2]:>10.1f}')
                print('-' *38)
            else:
                pass
    else:
        print('Codigo digitado inexistente!')
    print()
    opcao = str(input('Deseja consultar outro aluno? [S/N] ')).strip().upper()
print('-' * 32)
print('-------Sistema Finalizado!------')
print('-' * 32)
```
